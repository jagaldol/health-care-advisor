```python
import pandas as pd

# 전처리 끝난 final.csv
df = pd.read_csv('final.csv')
df.head()
```

```python
# Add a new column 'answer' to the DataFrame by concatenating 'answer_intro', 'answer_body', and 'answer_conclusion'
df['answer'] = df['answer_intro'] + " " + df['answer_body'] + " " + df['answer_conclusion']
df.head()
```

```python
import random

# 그룹화하고 각 그룹에서 랜덤하게 3개만 선택
df_sampled = df.groupby(['disease_category', 'disease_name_kor', 'department', 'intention']).apply(lambda x: x.sample(n=min(3, len(x)), random_state=42)).reset_index(drop=True)
df_sampled.head()
```

```python
len(df_sampled)
```

```python
# 60%: df의 answer 2개(disease_category	disease_name_kor	department	intention 전부 다른거)
# 40%: 자신의 answer + df의 answer 1개(disease_category, disease_name_kor, department, intention 전부 다른거)
df_sampled_40 = df_sampled.sample(frac=0.4, random_state=42).copy()
df_sampled_40['documents'] = df_sampled_40.apply(
    lambda row: [row['answer']] + df[
        (df['disease_category'] != row['disease_category']) &
        (df['disease_name_kor'] != row['disease_name_kor']) &
        (df['department'] != row['department']) &
        (df['intention'] != row['intention'])
    ]['answer'].sample(n=1, random_state=42).tolist(),
    axis=1
)

# 60%: df의 answer 2개(disease_category, disease_name_kor, department, intention 전부 다른거)
df_sampled_60 = df_sampled.drop(df_sampled_40.index).copy()
df_sampled_60['documents'] = df_sampled_60.apply(
    lambda row: df[
        (df['disease_category'] != row['disease_category']) &
        (df['disease_name_kor'] != row['disease_name_kor']) &
        (df['department'] != row['department']) &
        (df['intention'] != row['intention'])
    ]['answer'].sample(n=2, random_state=42).tolist(),
    axis=1
)

# 두 DataFrame을 다시 합침
df_sampled = pd.concat([df_sampled_40, df_sampled_60]).reset_index(drop=True)
df_sampled.head()
```

```python
df_sampled_clean = pd.read_csv('dataset.csv')
df_sampled_clean.head()
```

```python
# Drop the specified columns from the DataFrame
df_sampled_clean = df_sampled.drop(columns=['answer_intro', 'answer_body', 'answer_conclusion', 'Summary'])
df_sampled_clean = df_sampled_clean.rename(columns={'answer': 'gpt_document'})
df_sampled_clean['is_golden'] = df_sampled_clean.apply(lambda row: row['answer'] == row['documents'][0], axis=1)
df_sampled_clean.head()
```

```python
# 'match'가 True인 행들에 대해 'documents'의 순서를 뒤집음
df_sampled_clean.loc[df_sampled_clean['is_golden'] == True, 'documents'] = df_sampled_clean.loc[df_sampled_clean['is_golden'] == True, 'documents'].apply(lambda x: x[::-1] if random.random() < 0.5 else x)
df_sampled_clean.head()
```

```python
# Calculate the ratio of 'match' values
match_ratio = df_sampled_clean['is_golden'].value_counts(normalize=True)
print(match_ratio)
```

```python
df_sampled_clean.to_csv('dataset.csv', index=False)
```
